### Step 3: Generating Balanced Synthetic Reviews
To avoid class imbalance and overfitting, reviews for each travel personality are carefully crafted to ensure an equal distribution of `positive`, `neutral`, and `negative` sentiments. This ensures the dataset is robust for sentiment analysis and provides fair representation across all classes.

Steps:
1. Predefine the number of reviews per sentiment for each travel personality type.
2. Ensure an equal distribution of sentiments (`positive`, `neutral`, `negative`) for each class.
3. Assign synthetic reviews and their respective sentiments to the dataset.



```python
import pandas as pd
from collections import defaultdict
import random

# Load the dataset
file_path = "sentiment_analysis_data.csv"
travel_data = pd.read_csv(file_path)

# Review templates aligned with travel personalities
review_templates = {
    "Adventure Seeker": {
        "positive": [
            "Thrilling experience!", 
            "Loved every moment of the adventure.",
            "Can't wait for my next adrenaline-filled trip."
        ],
        "neutral": [
            "The adventure was okay, but not mind-blowing.", 
            "A decent experience overall.",
            "Met my expectations but didn't exceed them."
        ],
        "negative": [
            "Adventure spots were underwhelming.",
            "Expected more adrenaline rush than I got.",
            "Disappointed with the overall experience."
        ],
    },
    "Cultural Explorer": {
        "positive": [
            "Amazing cultural exposure!", 
            "Loved the historical tours and museums.",
            "Enriching experience filled with learning."
        ],
        "neutral": [
            "Some cultural spots were nice, others were average.", 
            "It was a mixed experience overall.",
            "I learned a lot but missed some expected highlights."
        ],
        "negative": [
            "Cultural landmarks were disappointing.", 
            "Expected more engaging activities.", 
            "Didn't feel the cultural immersion I hoped for."
        ],
    },
    "Luxury Traveler": {
        "positive": [
            "Fantastic luxury experience!", 
            "Loved the first-class treatment.", 
            "Everything felt so exclusive and elegant."
        ],
        "neutral": [
            "It was good but not worth the high price.", 
            "Decent amenities but lacked the wow factor.", 
            "Average luxury for the cost."
        ],
        "negative": [
            "Luxury standards were not met.", 
            "Felt overpriced for what was offered.", 
            "Disappointed with the quality of service."
        ],
    },
    "Balanced Traveler": {
        "positive": [
            "Great balance of adventure and relaxation.", 
            "Loved the mix of activities and downtime.", 
            "Perfect trip for my preferences."
        ],
        "neutral": [
            "The balance was okay, but not perfect.", 
            "I enjoyed some parts more than others.", 
            "Good trip overall but had room for improvement."
        ],
        "negative": [
            "Struggled to find a good mix of activities.", 
            "Felt unbalanced and poorly planned.", 
            "Didn't enjoy the trip as much as I hoped."
        ],
    },
}

# Track counts to ensure balance
sentiment_counts = defaultdict(lambda: {"positive": 0, "neutral": 0, "negative": 0})
max_reviews_per_sentiment = len(travel_data) // (len(review_templates) * 3)

# Function to assign balanced reviews
def balanced_review_assignment(personality_type):
    sentiments = ["positive", "neutral", "negative"]
    available_sentiments = [
        sentiment for sentiment in sentiments 
        if sentiment_counts[personality_type][sentiment] < max_reviews_per_sentiment
    ]
    
    # If all sentiment types are filled, randomly choose from any
    sentiment = random.choice(available_sentiments or sentiments)
    
    sentiment_counts[personality_type][sentiment] += 1
    return random.choice(review_templates[personality_type][sentiment]), sentiment

# Add synthetic reviews
reviews = []
sentiments = []
for index, row in travel_data.iterrows():
    personality = row["travel_personality"]
    review, sentiment = balanced_review_assignment(personality)
    reviews.append(review)
    sentiments.append(sentiment)

# Add reviews and sentiments to the dataset
travel_data["review"] = reviews
travel_data["sentiment"] = sentiments

# Save the enhanced dataset
enhanced_file_path = "enhanced_sentiment_analysis_data.csv"
travel_data.to_csv(enhanced_file_path, index=False)

# Display a preview of the enhanced data
print(travel_data.head())

```

### Step 5: Preprocess Text Data for Sentiment Analysis

The `review` column contains unstructured text data that must be preprocessed before training a machine learning model. This step involves:
1. Cleaning the text by removing unnecessary characters, punctuations, and whitespace.
2. Tokenizing the text into individual words.
3. Removing stop words and lemmatizing the text for standardization.

#### Objectives:
- Prepare the text data for vectorization and model training.
- Ensure the text data is clean, normalized, and suitable for NLP tasks.

#### Actions:
1. Perform text cleaning and preprocessing.
2. Save the processed reviews back into the dataset.
3. Verify the processed text data by printing sample reviews.

#### Expected Outcome:
A cleaned and preprocessed dataset ready for feature extraction and sentiment classification.


```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

# Function for text preprocessing
def preprocess_text(text):
    # Remove special characters and convert to lowercase
    text = re.sub(r"[^a-zA-Z\s]", "", text).lower()
    # Tokenize text
    tokens = word_tokenize(text)
    # Remove stopwords and lemmatize tokens
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    # Join tokens back into a single string
    return " ".join(tokens)

# Apply preprocessing to the review column
travel_data["processed_review"] = travel_data["review"].apply(preprocess_text)

# Preview the processed data
print("Sample Processed Reviews:")
print(travel_data[["review", "processed_review"]].head())
```

## Text Preprocessing for Sentiment Analysis

### Objective:
To clean and prepare the `review` column for sentiment analysis by removing noise, standardizing text, and extracting meaningful features.

### Process:
1. **Text Cleaning:**
   - Removed special characters and punctuation.
   - Converted text to lowercase for uniformity.

2. **Tokenization:**
   - Split sentences into individual words for analysis.

3. **Stopword Removal:**
   - Removed common words (e.g., "and", "the") using NLTK's English stopword list.

4. **Lemmatization:**
   - Reduced words to their base forms (e.g., "running" to "run") using WordNet Lemmatizer.

5. **Column Update:**
   - Created a new column `processed_review` to store the cleaned and processed reviews.

### Result:
The `processed_review` column now contains reviews in a cleaned format, with irrelevant noise removed and words standardized for better analytical performance.

## Review Vectorization for Sentiment Analysis

### Objective:
Convert the `processed_review` text data into numerical vectors suitable for machine learning models.

### Process:
1. **TF-IDF Vectorization:**
   - Applied Term Frequency-Inverse Document Frequency (TF-IDF) to represent the importance of words in each review relative to the entire dataset.
   - Limited the vocabulary to the top 1000 features to reduce dimensionality.

2. **Sparse Matrix Conversion:**
   - Transformed the vectorized data into a DataFrame for better interpretability and compatibility with machine learning models.

3. **Integration with Sentiment:**
   - Retained the `sentiment` column alongside the vectorized data for supervised learning.

### Result:
The reviews are now represented as numerical vectors, with each column corresponding to a unique word's importance score. This representation is ready for model training and evaluation.

### Example:
| **Word Features**      | **...** | **Sentiment** |
|-------------------------|---------|---------------|
| 0.123 (importance of word 1) | ...     | Negative      |
| 0.045 (importance of word 2) | ...     | Positive      |

This step ensures the text data is numerically encoded for effective training with machine learning algorithms.


```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Adjust max_features for dimensionality reduction

# Apply vectorization to the processed reviews
review_vectors = tfidf_vectorizer.fit_transform(travel_data["processed_review"])

# Convert the sparse matrix to a DataFrame for better interpretability
vectorized_reviews = pd.DataFrame(
    review_vectors.toarray(), 
    columns=tfidf_vectorizer.get_feature_names_out()
)

# Add sentiment column to the vectorized data for model training
vectorized_reviews["sentiment"] = travel_data["sentiment"]

# Preview the vectorized data
print("Vectorized Review Data:")
print(vectorized_reviews.head())
```

### Step 7.1: Train-Test Split and Model Selection

#### **Objective:**
To build machine learning models that predict the sentiment (positive, neutral, or negative) based on the vectorized review data.

#### **Steps Taken:**
1. **Split the Data:**
   - Divided the dataset into training and testing sets to evaluate the model's generalization.
   - Used an 80-20 split for training and testing data.

2. **Model Selection:**
   - Chose commonly used classifiers for text-based sentiment analysis:
     - Logistic Regression
     - Random Forest
     - Gradient Boosting (e.g., XGBoost)

3. **Evaluation Metrics:**
   - Accuracy: Proportion of correct predictions out of total predictions.
   - F1-score: Balance between precision and recall, especially for imbalanced datasets.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Separate features and target
X = vectorized_reviews.drop(columns=['sentiment'])  # Vectorized review features
y = vectorized_reviews['sentiment']  # Sentiment labels

# Encode target labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train and evaluate Logistic Regression model
lr_model = LogisticRegression(max_iter=200, random_state=42)
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)

print("Logistic Regression Results:")
print(f"Accuracy: {accuracy_score(y_test, lr_predictions):.2f}")
print("Classification Report:")
print(classification_report(y_test, lr_predictions, target_names=label_encoder.classes_))
```
